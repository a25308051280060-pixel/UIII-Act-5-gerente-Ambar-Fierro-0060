#### gerente y planificacion


#### El gerente es el pilar central de una organización, responsable de coordinar equipos, optimizar recursos y ejecutar estrategias. Su objetivo principal es asegurar que los objetivos operativos y comerciales se cumplan de manera eficiente, guiando a la empresa hacia el éxito empresarial.Las funciones principales de un gerente se dividen en cuatro pilares fundamentales:1. PlanificaciónDefinición de objetivos: Establece metas a corto, mediano y largo plazo.Planeación estratégica: Diseña los planes de acción y presupuestos necesarios para materializar la visión de la empresa.

```python
# --- SIMULADOR DE IA PARA PLANIFICACIÓN GERENCIAL v1.0 ---

print("Bienvenido, Gerente. Soy su Asistente de IA de Planificación.")
print("Por favor, responda a las siguientes preguntas para procesar su estrategia.\n")

# Entrada interactiva de datos (Interactividad) [5]
nombre_proyecto = input("Nombre del proyecto: ")
presupuesto = float(input("Presupuesto asignado ($): "))
prioridad = int(input("Nivel de prioridad (1 al 10): "))
miembros_equipo = int(input("Cantidad de personal disponible: "))
es_urgente = input("¿Es de entrega inmediata? (si/no): ")

# Procesamiento de variables lógicas para la IA
urgencia_logica = es_urgente == "si" # Operador relacional de igualdad [3]

# --- LÓGICA DE DECISIÓN DE LA IA ---

print("\n--- ANALIZANDO DATOS Y GENERANDO PLANIFICACIÓN ---")

# Cálculo de un 'Índice de Viabilidad' mediante operadores aritméticos [6, 7]
# El presupuesto por persona ayuda a determinar la robustez del plan
presupuesto_por_persona = presupuesto / miembros_equipo

# CASO 1: Planificación de Emergencia (Alta Urgencia y Alta Prioridad)
# Utiliza operadores lógicos 'and' y relacionales '>' [3, 4]
if urgencia_logica and prioridad > 8:
    estrategia = "DESPLIEGE INMEDIATO Y MÁXIMA PRIORIDAD"
    costo_con_contingencia = presupuesto * 1.20 # Aritmético: 20% extra por urgencia
    print(f"IA Detectó: Estado Crítico para el proyecto '{nombre_proyecto}'.")
    print(f"Acción Gerencial: {estrategia}.")
    print(f"Planificación: Se autoriza un presupuesto de contingencia de ${costo_con_contingencia:.2f}.")

# CASO 2: Optimización de Recursos (Presupuesto Ajustado pero importante)
# Utiliza operadores lógicos 'or' y 'and' combinados [8]
elif presupuesto < 5000 or (prioridad >= 5 and prioridad <= 8):
    estrategia = "OPTIMIZACIÓN DE RECURSOS Y FASES ESCALONADAS"
    print(f"IA Detectó: El proyecto '{nombre_proyecto}' requiere manejo eficiente.")

    # Condicional anidado para evaluar el equipo [9]
    if miembros_equipo < 3:
        print("Nota de IA: Personal insuficiente. Se recomienda contratar consultoría externa.")
    else:
        print(f"Acción Gerencial: {estrategia}.")
        print(f"Planificación: Iniciar fase 1 con {miembros_equipo} empleados.")

# CASO 3: Proyectos de Bajo Impacto o Falta de Viabilidad Económica
# Utiliza el operador lógico 'not' [10]
elif not urgencia_logica and presupuesto_por_persona < 500:
    estrategia = "POSTERGACIÓN O REEVALUACIÓN DE COSTOS"
    print(f"IA Detectó: Viabilidad económica baja (${presupuesto_por_persona} por persona).")
    print(f"Acción Gerencial: {estrategia}.")
    print("Planificación: El proyecto queda en espera de mayor presupuesto.")

# CASO 4: Caso por defecto para situaciones no categorizadas [11]
else:
    print(f"IA Detectó: El proyecto '{nombre_proyecto}' tiene parámetros fuera de los modelos estándar.")
    print("Acción Gerencial: REVISIÓN MANUAL NECESARIA.")
    print("Planificación: La IA no puede decidir automáticamente; requiere juicio humano.")

print("\n--- PROCESO DE PLANIFICACIÓN FINALIZADO ---")
print("programa elaborado por Ambar Fierro NL = 19 gerente")
```
